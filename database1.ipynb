```python
import matplotlib.pyplot as plt

# Créer une figure et un axe
fig, ax = plt.subplots(figsize=(12, 6))

# Calculer la population mondiale par année en faisant la somme sur toutes les lignes
world_population = df.iloc[:, 1:].sum()

# Créer un graphique à barres horizontales (histogramme)
ax.bar(world_population.index, world_population.values)

# Ajouter des titres et des étiquettes
ax.set_title('Évolution de la population mondiale (1960-2022)')
ax.set_xlabel('Année')
ax.set_ylabel('Population mondiale')

# Faire pivoter les étiquettes d'année pour une meilleure lisibilité
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt

data = df.set_index('Country Name').loc[:, '1960':'2022']

data = data.T

# Créer un graphique pour chaque pays
for country in data.columns:
    plt.figure(figsize=(12, 6))
    plt.bar(data.index, data[country])
    plt.title(f'Population de {country} (1960-2022)')
    plt.xlabel('Année')
    plt.ylabel('Population')
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()
```
